# Read in data to sort

```python
import json
import GenomeAligner
import random

with open('testData.json', 'r') as f:
    data = json.load(f)

newest_individual = data[-1]
random.shuffle(data)
```

```python
# Create a sequence aligner object
genomeLength = len(data[0][1])
ga = GenomeAligner.GenomeAligner(genomeLength, genomeLength)

sortedList = []
current_closest = newest_individual

while current_closest != None:
    sortedList.append(current_closest)
    print(f"Removing {current_closest}")
    data.remove(current_closest)
    current_closest = ga.find_best_match(current_closest, data)
```

```python
sorted_data = []
for ind in sortedList:
    copied_individual = []
    for item in ind:
        copied_individual.append(item)
    sorted_data.append(copied_individual)

sorted_data = sorted_data[::-1]
print('[Real Placement, Decoded Genome, Fitness]')
for i in range(0, len(sorted_data)):
    distance_from_actual = sorted_data[i][0]-i
    if distance_from_actual < 0:
        distance_from_actual *= -1
    sorted_data[i].append(distance_from_actual)
    print(sorted_data[i])

# print(sortedList)

# for i in range(0, len(data)):
#     print(data[i][0], data[i][3])
# data.sort(key= lambda x:x[3])
# for i in range(0, len(data)):
#     data[i].append(i*3)
#     print(int(data[i][4])-int(data[i][0]), data[i][3])


def col_average(arr, column):
    total = 0
    for i in arr:
        total += i[column]
        # diff = i[4] - i[0]
        # total += diff if diff > 0 else -1 * diff
    return total / len(arr)

print(col_average(sorted_data, 3))
```

```python
outputData = []
with open("sortedResults1.csv", "w") as outfile:
    outfile.write('Actual Location,Genome,Fitness,DistanceFromActual\n')
    for ind in sorted_data:
        outfile.write(f'{ind[0]},{ind[1]},{ind[2]},{ind[3]}\n')
```
